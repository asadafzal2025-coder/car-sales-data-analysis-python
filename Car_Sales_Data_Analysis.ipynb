```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
brand_count = car_sales["Brand"].value_counts()

plt.bar(brand_count.index, brand_count.values)
plt.title("Number of Cars by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

###
Histogram — Price distribution


```python
plt.hist(car_sales['Price'], bins=20)
plt.title("Price vs Frequency")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

###

```python
plt.scatter(car_sales['Mileage'], car_sales['Price'])
plt.title("Mileage vs Price")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

###

# Executive Summary

This project analyzes the car sales dataset to identify key trends in vehicle inventory, pricing, and mileage. The objective was to provide actionable insights that can support better inventory management and pricing decisions.

## Key Findings

* **Most Common Manufacturer:** Volkswagen had the highest number of vehicles in the inventory, indicating that it is the dealership's most frequently stocked brand.
* **Highest Average Selling Price:** Mercedes-Benz recorded the highest average selling price among all manufacturers, suggesting that it occupies the premium segment of the inventory.
* **Average Mileage:** The average mileage of vehicles in the dataset was 161.237 km, representing the typical usage level of the available vehicles.
* **Missing Data:** Missing values were identified during data exploration and were handled using appropriate data-cleaning techniques to ensure accurate analysis.

## Business Insights

* Premium brands such as Mercedes-Benz command significantly higher average selling prices than economy brands.
* Toyota dominates the inventory, which may reflect strong customer demand and reliable market performance.
* Vehicles with lower mileage generally tend to have higher selling prices, indicating a relationship between vehicle usage and market value.
* The dealership maintains a diverse inventory across multiple body types, allowing it to serve a wide range of customer preferences.

## Recommendations

1. Increase inventory of high-demand manufacturers while maintaining a balanced selection of premium vehicles.
2. Promote low-mileage vehicles in marketing campaigns, as they typically attract higher selling prices.
3. Continue monitoring pricing trends by manufacturer and body type to optimize profitability.
4. Implement regular data-quality checks to minimize missing values and improve the reliability of future analyses.

## Conclusion

The analysis provides valuable insights into the dealership's inventory composition, pricing structure, and vehicle characteristics. These findings can support data-driven decision-making, improve inventory planning, and enhance overall business performance.
